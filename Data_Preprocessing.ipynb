# Analyse Dataset Attributes

```python
import h5py

file_path = "./workspaces/hdf5s/smd/Bach_BWV849-01_001_20090916-SMD.h5"

with h5py.File(file_path, 'r') as hf:
    # List all attributes
    print("SMD Attributes:", list(hf.attrs.keys()))

    # Access the 'duration' attribute
    duration = hf.attrs['duration']
    print("Duration:", duration)

    # Access the 'duration' attribute
    midi_filename = hf.attrs['midi_filename']
    print("MIDI:", midi_filename)

    # Access the 'duration' attribute
    audio_filename = hf.attrs['audio_filename']
    print("Audio:", audio_filename)

    # Access the 'duration' attribute
    split = hf.attrs['split']
    print("Split:", split)
```
